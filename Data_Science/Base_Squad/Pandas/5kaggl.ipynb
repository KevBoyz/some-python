```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./assets2/student_habits_performance.csv')
df.head()
```

```python
df.describe()
```

```python
scatter = plt.scatter(
    df['study_hours_per_day'], 
    df['exam_score'], 
    c=df['mental_health_rating'], cmap='Blues',
    alpha=0.9)

plt.colorbar(scatter, label='Mental Health')
plt.xlabel('Study time  (Hours)')
plt.ylabel('Exam Score')
plt.grid(alpha=0.2)
plt.show()
```

```python
scatter = plt.scatter(
    df['part_time_job'], 
    df['exam_score'], 
    c=df['mental_health_rating'], cmap='Blues',
    alpha=0.9)

plt.colorbar(scatter, label='Mental Health')
plt.xlabel('Pat time job')
plt.ylabel('Exam Score')
plt.grid(alpha=0.2)
plt.show()
```

```python
import pandas as pd  # Gerado com ia
import matplotlib.pyplot as plt
import numpy as np # Importamos numpy para usar o jitter

# Supondo que seu DataFrame 'df' j esteja carregado

plt.figure(figsize=(8, 6)) # Ajustando o tamanho da figura para melhor visualizao

# Adicionando um pequeno jitter no eixo x para a varivel categrica 'part_time_job'
jitter = 0.1
x_jitter_no = np.random.uniform(-jitter, jitter, size=len(df[df['part_time_job'] == 'No']))
x_no = np.zeros(len(df[df['part_time_job'] == 'No'])) + x_jitter_no

x_jitter_yes = np.random.uniform(-jitter, jitter, size=len(df[df['part_time_job'] == 'Yes']))
x_yes = np.ones(len(df[df['part_time_job'] == 'Yes'])) + x_jitter_yes

scatter_no = plt.scatter(x_no, df.loc[df['part_time_job'] == 'No', 'exam_score'],
                         c=df.loc[df['part_time_job'] == 'No', 'mental_health_rating'],
                         cmap='Blues', alpha=0.8, label='No Part-Time Job')

scatter_yes = plt.scatter(x_yes, df.loc[df['part_time_job'] == 'Yes', 'exam_score'],
                          c=df.loc[df['part_time_job'] == 'Yes', 'mental_health_rating'],
                          cmap='Blues', alpha=0.8, label='Has Part-Time Job')

cbar = plt.colorbar(scatter_yes, label='Mental Health Rating') # Usamos scatter_yes para a colorbar

plt.xlabel('Part-Time Job')
plt.ylabel('Exam Score')
plt.title('Exam Score vs. Part-Time Job Colored by Mental Health Rating')
plt.xticks([0, 1], ['No', 'Yes']) # Garantindo que os ticks do eixo x sejam 'No' e 'Yes'
plt.grid(alpha=0.3) # Tornando a grade mais sutil
plt.legend() # Adicionando legenda para os grupos (opcional, mas pode ser til)
plt.tight_layout() # Ajusta o layout para evitar cortes
plt.show()
```
