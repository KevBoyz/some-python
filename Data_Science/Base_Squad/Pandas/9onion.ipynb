```python
import pandas as pd
```

```python
# Create first DataFrame: Customer Information
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'age': [25, 30, 35, 40, 28],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Boston']
})

# Create second DataFrame: Orders
orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105, 106],
    'customer_id': [1, 2, 2, 3, 5, 6],  # Note: customer_id 6 has no match in customers
    'product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard', 'Mouse'],
    'amount': [1000, 800, 300, 250, 100, 50]
})
```

```python
def fun(group):
    return group - 1000


df = pd.merge(customers, orders, on='customer_id', how='left')
df['amount'].apply(fun)
```
