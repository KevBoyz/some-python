```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
%matplotlib inline
```

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'], 'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
df = pd.DataFrame(data=d) 
df.head()
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
ohe_transformed = ohe.fit_transform(df[['city']])
```

```python
tr_df = pd.concat([df, ohe_transformed], axis=1).drop(columns=['city'])
```

```python
tr_df
```

```python
categories = [i for i in tr_df['size'].unique()]

# NÃºmero com base na ordem da lista
oe = OrdinalEncoder(handle_unknown='error', categories=[['Large', 'Small', 'Medium']])
tr_df['size'] = oe.fit_transform(tr_df[['size']])
```
