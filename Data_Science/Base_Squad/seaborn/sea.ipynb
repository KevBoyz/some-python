```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg
```

```python
sns.set_theme()
sns.set_style('darkgrid') 
#sns.set_context('paper', font_scale=1.2)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=mpg, x='acceleration', y='mpg', hue='model_year', palette='rainbow')
```

```python
sns.lineplot(mpg, x='model_year', y='mpg')
```

```python
sns.pointplot(mpg, x='model_year', y='mpg')
```

```python
sns.kdeplot(data=mpg, x='acceleration', y='mpg')
```

```python
sns.kdeplot(data=mpg, x='acceleration', y='mpg', fill=True)
```

```python
sns.histplot(data=mpg, x='acceleration', y='mpg', fill=True)
```

```python
sns.histplot(mpg, x='mpg')
```

```python
corr = mpg.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```
