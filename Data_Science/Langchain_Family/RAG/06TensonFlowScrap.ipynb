```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://www.tensorflow.org'
```

```python
soup = BeautifulSoup(
    requests.get('https://www.tensorflow.org/guide').text,
    'html.parser'
)
```

```python
navbar = soup.find('nav', class_='devsite-book-nav devsite-nav nocontent')
div = navbar.find('div', class_='devsite-mobile-nav-bottom') 
ul = div.find('ul')
anchors = ul.find_all('a')

urls = [base_url + a.get('href') for a in anchors[1:]]
```

```python
for a in urls:
    print(a)
```
